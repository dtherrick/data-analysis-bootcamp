```python
dfTitanicData['GenderLongWay'] = 4
dfTitanicData['GenderLongWay'] = dfTitanicData['Sex'].map({'female': 0, 'male':1}).astype(int)
dfTitanicData.head(3)
```

```python
dfTitanicData['GenderLongWay'].value_counts()
```

```python
dfTitanicData.info()
```

```python
dfTitanicData.dropna(how='any', axis=0, subset=['Embarked'], inplace=True)
```

```python
dfTitanicData['EmbarkedLongWay'] = 999
dfTitanicData['EmbarkedLongWay'] = dfTitanicData['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
```

```python
dfTitanicData['GenderShortWay'] = pd.Categorical(dfTitanicData['Sex']).codes
dfTitanicData['EmbarkedShortWay'] = pd.Categorical(dfTitanicData['Embarked']).codes
```

```python
dfTitanicData[['Sex', 'GenderLongWay', 'GenderShortWay', 'Embarked', 'EmbarkedLongWay', 'EmbarkedShortWay']]
```

```python
median_ages = np.zeros((2,3))
median_ages
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = dfTitanicData[(dfTitanicData['GenderLongWay'] == i) & (dfTitanicData['Pclass'] == j+1)]['Age'].dropna().median()
        
median_ages
```

```python
dfTitanicData['AgeFill'] = dfTitanicData['Age']
```

```python
dfTitanicData[dfTitanicData['Age'].isnull()][['GenderLongWay', 'Pclass', 'Age', 'AgeFill']].head(10)
```

```python
for i in range(0,2):
    for j in range(0,3):
        dfTitanicData.loc[(dfTitanicData.Age.isnull()) & (dfTitanicData['GenderLongWay'] == i) & (dfTitanicData['Pclass'] == j+1), 'AgeFill'] = median_ages[i,j]
```

```python
dfTitanicData[dfTitanicData['Age'].isnull()][['GenderLongWay', 'Pclass', 'Age', 'AgeFill']].head(10)
```

```python
dfTitanicData['AgeIsNull'] = pd.isnull(dfTitanicData.Age).astype(int)
```

```python
dfTitanicData.describe()
```

```python
dfTitanicData['AgeFill'].hist(bins=16, range=(0,80), alpha=0.5)
plt.show()
```

```python
dfTitanicData['FamilySize'] = dfTitanicData['SibSp'] + dfTitanicData['Parch']
dfTitanicData['AgeXPclass'] = dfTitanicData.AgeFill * dfTitanicData.Pclass
```
